```python
"""
STATIC METHOD
They are somehow related to Class but does not depend of Class or isntance 
It is simply a function whose first variable is neither class nor a instnace
"""
class Employee:
    
    raise_amt= 1.30
    no_of_emps = 0
    def __init__(self,first,last,pay):
        self.first = first    #fisrt,last,pay,email all are attributes
        self.last = last
        self.pay = pay
        self.email = first+'.'+last+'@company.com'
        Employee.no_of_emps +=1
        
    def fullname(self):
        return '{} {}'.format(self.first,self.last)#fullname is a method

    def apply_raise(self):
        self.pay = int(self.pay*self.raise_amt) #or self.pay = int(self.pay*Employee.raise_amount) 
        return(self.pay)  # 4% increment
    
    @classmethod
    def set_raise_amt(cls,amount):
        cls.raise_amt=amount
        
    @classmethod   #Alternate class methos for parsing the string
    def from_string(cls,emp_str):
        first,last,pay = emp_str.split('-')
        return cls(first,last,pay)
    
    @staticmethod
    
    def is_workday(day):
        if day.weekday()==5 or  day.weekday()==6:
            return 'Not a workday'
        return 'Workday'
    
        
    
    

emp1 = Employee('Utkarsh','ujwal',1000000)
emp2 = Employee('Ankit','Jha',1100000)

import datetime

my_date= datetime.date(2019,8,20)

print(Employee.is_workday(my_date))
```

```python
"""
INHERITANCE : CREATING SUBCLASS

"""
class Employee:
    
    raise_amt= 1.04
    
    def __init__(self,first,last,pay):
        self.first = first    #fisrt,last,pay,email all are attributes
        self.last = last
        self.pay = pay
        self.email = first+'.'+last+'@company.com'
        
        
    def fullname(self):
        return '{} {}'.format(self.first,self.last)#fullname is a method

    def apply_raise(self):
        self.pay = int(self.pay*self.raise_amt) #or self.pay = int(self.pay*Employee.raise_amount) 
        return(self.pay)  # 4% increment
    
class Developer(Employee):
    raise_amt=1.10 #suppose we want out developers to get 10% raise


#dev1= Employee('Utkarsh','ujwal',1000000)
#dev2 = Employee('Ankit','Jha',1100000)

dev_1= Developer('Utkarsh','ujwal',1000000)
dev_2 = Developer('Ankit','Jha',1100000)


#print(dev_1.pay)
#dev_1.apply_raise()
#print(dev_1.pay)

#print(dev1.email)
#print(dev2.email)
#print('\n')
#print(dev_1.email)
#print(dev_2.email)
#print('\n')
#print(help(Developer))

"""
Sometimes we want our subclass to be intiated with more information than the parent class
can handle, so we need to give the subclass its own __init__ method
Here we want to give developer an option to add programming language too
"""

class Employee:
    
    raise_amt= 1.04
    
    def __init__(self,first,last,pay):
        self.first = first    #fisrt,last,pay,email all are attributes
        self.last = last
        self.pay = pay
        self.email = first+'.'+last+'@company.com'
        
        
    def fullname(self):
        return '{} {}'.format(self.first,self.last)#fullname is a method

    def apply_raise(self):
        self.pay = int(self.pay*self.raise_amt) #or self.pay = int(self.pay*Employee.raise_amount) 
        return(self.pay)  # 4% increment
    
class Developer(Employee):
    raise_amt=1.10 #suppose we want out developers to get 10% raise
    
    def __init__(self,first,last,pay,prog_lang):
        super().__init__(first,last,pay) #instead to writing the entire code we use super to let Employee handle those things
        # Employee.__init__(self,first,last,pay)  #to call parent's init method
        self.prog_lang = prog_lang
        
class Manager(Employee):
    def __init__(self,first,last,pay,employee=None):
        super().__init__(first,last,pay) #instead to writing the entire code we use super to let Employee handle those things
        # Employee.__init__(self,first,last,pay)  #to call parent's init method
        if employee==None:
            employee=[]
        else:
            self.employee= employee
#List of employee the manager supervises
    def add_emp(self,emp):
        if emp not in self.employee:
            self.employee.append(emp)
            
    def remove_emp(self,emp):
        if employee in self.employee:
            self.employee.remove(emp)
            
    #print employee list 
    
    def print_emp(self):
        for emp in self.employee:
            print("---->", emp.fullname())

            
            
#dev1= Employee('Utkarsh','ujwal',1000000)
#dev2 = Employee('Ankit','Jha',1100000)

dev_1= Developer('Utkarsh','ujwal',1000000, 'Python')
dev_2 = Developer('Ankit','Jha',1100000,'Java')

mgr_1 = Manager('Susan','Jain',900000, [dev_1])

print(mgr_1.email)
print(mgr_1.add_emp(dev_2))
print(mgr_1.print_emp())
#print(dev_1.pay)
#print(dev_1.prog_lang)
#print(dev_2.pay)
#print(dev_2.prog_lang)
```

```python
"""

Special (magic/Dunder) method
Those are __repr__ and __str__  to represent the instances in a readable format
They methods allow to change how our objects are displayed
"""

class Employee:
    
    raise_amt= 1.04
    
    def __init__(self,first,last,pay):
        self.first = first    #fisrt,last,pay,email all are attributes
        self.last = last
        self.pay = pay
        self.email = first+'.'+last+'@company.com'
        
        
    def fullname(self):
        return '{} {}'.format(self.first,self.last)#fullname is a method

    def apply_raise(self):
        self.pay = int(self.pay*self.raise_amt) #or self.pay = int(self.pay*Employee.raise_amount) 
        return(self.pay)  # 4% increment
    
    def __repr__(self):
        return "Employee( '{}','{}','{}'))".format(self.first,self.last,self.pay)
    
    def __str__(self):
         return '{} - {}'.format(self.fullname(),self.email)
    
emp1 = Employee('Utkarsh','ujwal',1000000)
emp2 = Employee('Ankit','Jha',1100000)
                                                   
#print(emp1) 
print(repr(emp1))
print(str(emp1))
                                                   
                                                   
                                                   
                                                   
```
